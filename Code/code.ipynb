Merging all the data together

```python
import pandas as pd
import os
import re

# Function to clean currency strings (remove ₹, $, commas, etc.)
def clean_currency(value):
    if pd.isna(value) or value == "":
        return 0.0
    value = str(value).strip()
    value = re.sub(r'[₹$,]', '', value).replace(',', '')
    try:
        return float(value)
    except ValueError:
        return 0.0

# Define the input Excel file paths
input_files = [
    r"C:\Users\SHAHIL\Downloads\INTERN PROJECT\t1\OUTPUT\Amazon_Order_113-6135300.xlsx",
    r"C:\Users\SHAHIL\Downloads\INTERN PROJECT\t1\OUTPUT\Amazon_Order_407-3521143-2456518.xlsx",
    r"C:\Users\SHAHIL\Downloads\INTERN PROJECT\t1\OUTPUT\Flipkart_Order_OD30418030310.xlsx",
    r"C:\Users\SHAHIL\Downloads\INTERN PROJECT\t1\OUTPUT\Flipkart_Order_OD119964200686387000.xlsx"
]

# Define the mandatory accounting columns
accounting_columns = [
    "Order ID",
    "Order Date",
    "Invoice Number",
    "Invoice Date",
    "Billing Address",
    "Product Description",
    "Quantity",
    "Unit Price",
    "Tax Amount",
    "Total Amount",
    "Grand Total"
]

# Initialize an empty list to store DataFrames
dfs = []

# Process each Excel file
for file in input_files:
    if not os.path.exists(file):
        print(f"Warning: File {file} not found. Skipping.")
        continue
    
    try:
        df = pd.read_excel(file, sheet_name="Order Details")
    except Exception as e:
        print(f"Warning: Error reading {file}: {e}. Skipping.")
        continue
    
    # Initialize mapped DataFrame with default empty strings
    mapped_df = pd.DataFrame({
        col: [""] * len(df) if col != "Quantity" else [1] * len(df)
        for col in accounting_columns
    })
    
    # Map columns
    if "Order ID" in df.columns:
        mapped_df["Order ID"] = df["Order ID"].astype(str)
    elif "Order Number" in df.columns:
        mapped_df["Order ID"] = df["Order Number"].astype(str)
    
    if "Order Date" in df.columns:
        mapped_df["Order Date"] = df["Order Date"].astype(str)
    
    if "Invoice Number" in df.columns:
        mapped_df["Invoice Number"] = df["Invoice Number"].astype(str)
    elif "Invoice No" in df.columns:
        mapped_df["Invoice Number"] = df["Invoice No"].astype(str)
    
    if "Invoice Date" in df.columns:
        mapped_df["Invoice Date"] = df["Invoice Date"].astype(str)
    
    if "Billing Address" in df.columns:
        mapped_df["Billing Address"] = df["Billing Address"].astype(str)
    
    # Handle product descriptions and quantities
    if "Item 1 Description" in df.columns:
        descriptions = []
        quantities = []
        unit_prices = []
        tax_amounts = []
        total_amounts = []
        
        if "Item 1 Description" in df.columns and pd.notna(df["Item 1 Description"].iloc[0]):
            descriptions.append(str(df["Item 1 Description"].iloc[0]))
            quantities.append(str(df["Item 1 Quantity"].iloc[0]))
            unit_prices.append(str(df["Item 1 Unit Price"].iloc[0]))
            tax_amounts.append(str(df["Item 1 Tax Amount"].iloc[0]))
            total_amounts.append(str(df["Item 1 Total"].iloc[0]))
        
        if "Item 2 Description" in df.columns and pd.notna(df["Item 2 Description"].iloc[0]):
            descriptions.append(str(df["Item 2 Description"].iloc[0]))
            quantities.append(str(df["Item 2 Quantity"].iloc[0]))
            unit_prices.append(str(df["Item 2 Unit Price"].iloc[0]))
            tax_amounts.append(str(df["Item 2 Tax Amount"].iloc[0]))
            total_amounts.append(str(df["Item 2 Total"].iloc[0]))
        
        mapped_df["Product Description"] = "; ".join(descriptions)
        mapped_df["Quantity"] = "; ".join(quantities)
        mapped_df["Unit Price"] = "; ".join(unit_prices)
        mapped_df["Tax Amount"] = "; ".join(tax_amounts)
        mapped_df["Total Amount"] = "; ".join(total_amounts)
    else:
        if "Product Description" in df.columns:
            mapped_df["Product Description"] = df["Product Description"].astype(str)
        elif "Item Description" in df.columns:
            mapped_df["Product Description"] = df["Item Description"].astype(str)
        
        if "Quantity" in df.columns:
            mapped_df["Quantity"] = df["Quantity"].astype(str)
        
        if "Unit Price" in df.columns:
            mapped_df["Unit Price"] = df["Unit Price"].astype(str)
        elif "Item Price" in df.columns:
            mapped_df["Unit Price"] = df["Item Price"].astype(str)
        
        if "Tax Amount" in df.columns:
            mapped_df["Tax Amount"] = df["Tax Amount"].astype(str)
        elif "Sales Tax" in df.columns:
            mapped_df["Tax Amount"] = df["Sales Tax"].astype(str)
        elif "Total Tax Amount" in df.columns:
            mapped_df["Tax Amount"] = df["Total Tax Amount"].astype(str)
        
        if "Total Amount" in df.columns:
            mapped_df["Total Amount"] = df["Total Amount"].astype(str)
        elif "Item 1 Total" in df.columns:
            mapped_df["Total Amount"] = df["Item 1 Total"].astype(str)
        elif "Item Price" in df.columns:
            mapped_df["Total Amount"] = df["Item Price"].astype(str)
    
    # Handle tax amounts for Flipkart_2 (CGST + SGST)
    if "CGST Amount" in df.columns and "SGST/UTGST Amount" in df.columns:
        cgst = df["CGST Amount"].apply(clean_currency)
        sgst = df["SGST/UTGST Amount"].apply(clean_currency)
        mapped_df["Tax Amount"] = (cgst + sgst).astype(str)
    
    if "Grand Total" in df.columns:
        mapped_df["Grand Total"] = df["Grand Total"].astype(str)
    
    # Fill any NaN values
    mapped_df = mapped_df.fillna({"Quantity": "1", "Tax Amount": "0.0"})
    mapped_df = mapped_df.fillna("")
    
    dfs.append(mapped_df)

# Concatenate all DataFrames
if dfs:
    merged_df = pd.concat(dfs, ignore_index=True)
    
    # Clean currency fields
    merged_df["Unit Price"] = merged_df["Unit Price"].apply(clean_currency)
    merged_df["Tax Amount"] = merged_df["Tax Amount"].apply(clean_currency)
    merged_df["Total Amount"] = merged_df["Total Amount"].apply(clean_currency)
    merged_df["Grand Total"] = merged_df["Grand Total"].apply(clean_currency)
    
    # Ensure all columns are present
    for col in accounting_columns:
        if col not in merged_df.columns:
            merged_df[col] = "" if col != "Quantity" else "1"
    
    # Reorder columns
    merged_df = merged_df[accounting_columns]
    
    # Define the output Excel file path
    output_path = r"C:\Users\SHAHIL\Downloads\INTERN PROJECT\t1\OUTPUT\Merged_Accounting_Orders.xlsx"
    
    try:
        merged_df.to_excel(output_path, index=False, sheet_name="Accounting Details")
        print(f"Merged Excel file has been created successfully at: {output_path}")
    except Exception as e:
        print(f"Error saving Excel file: {e}")
else:
    print("No valid Excel files were found to merge.")
```
